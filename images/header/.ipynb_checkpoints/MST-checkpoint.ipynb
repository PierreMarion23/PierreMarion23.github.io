# Minimum Spanning Tree generator

```python
from matplotlib import pyplot as plt
import matplotlib
import numpy as np
import scipy.sparse
```

```python
N = 150
```

```python
X, Y = np.random.rand(N), np.random.rand(N)
M = np.zeros((N, N))
for i in range(N):
    for j in range(N):
        M[i, j] = np.sqrt((X[i]-X[j])**2 + (Y[i]-Y[j])**2)
```

```python
sp_tree = scipy.sparse.csgraph.minimum_spanning_tree(M)
```

```python
liaisons = sp_tree.nonzero()
```

```python
import cycler

color = plt.cm.inferno(np.linspace(0, 1,N))
matplotlib.rcParams['axes.prop_cycle'] = cycler.cycler('color', color)
```

```python
dpi = 100
plt.figure(figsize=(2000/dpi, 200/dpi), dpi=dpi)
plt.subplots_adjust(left=0, bottom=0, right=1, top=1, wspace=None, hspace=None)
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.axis('off')
plt.scatter(X, Y)
for k in range(N-1):
    if np.random.rand() < 0.9:
        plt.plot([X[liaisons[0][k]], X[liaisons[1][k]]], [Y[liaisons[0][k]], Y[liaisons[1][k]]])
plt.savefig("header.svg", format="svg")
plt.show()
```
